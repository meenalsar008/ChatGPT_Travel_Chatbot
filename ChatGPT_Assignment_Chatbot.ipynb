```python
import os
import signal

def find_and_kill_process_by_port(port):
    try:
        pid = None

        # Find the process using the specified port
        cmd = f"lsof -i :{port}"
        process_info = os.popen(cmd).read()

        # Extract the process ID (PID)
        lines = process_info.strip().split('\n')
        if len(lines) > 1:
            pid = int(lines[1].split()[1])

        if pid:
            # Terminate the process
            os.kill(pid, signal.SIGTERM)
            print(f"Process with PID {pid} on port {port} has been terminated.")
        else:
            print(f"No process found on port {port}.")
    except Exception as e:
        print(f"Error: {str(e)}")

# Specify the port number you want to stop
port_to_stop = 5000
find_and_kill_process_by_port(port_to_stop)

port_to_stop_2 = 5001
find_and_kill_process_by_port(port_to_stop_2)

port_to_stop_3 = 5050
find_and_kill_process_by_port(port_to_stop_3)
```

```python
# database

from flask import Flask, request, jsonify
import sqlite3
import threading

app = Flask(__name__)

# Function to connect to the database
def connect_db():
    return sqlite3.connect('travel_agent_database.db')

# Endpoint to execute a custom query
@app.route('/query', methods=['GET'])
def execute_query():
    try:
        query = request.args.get('query')

        conn = connect_db()
        cursor = conn.cursor()
        cursor.execute(query)
        result = cursor.fetchall()
        column_names = [description[0] for description in cursor.description]
        conn.close()

        return jsonify({'column_names': column_names, 'result': result})
    except Exception as e:
        return jsonify({'error': str(e)}), 500

# Create a separate thread for running the Flask app
def run_app():
    app.run(host='0.0.0.0', port=5050)

if __name__ == '__main__':
    t2 = threading.Thread(target=run_app)
    t2.start()
```

```python
import requests

def create_table_definition_prompt(api_url, query):
    try:
        # Send the query to your API
        response = requests.get(api_url, params={"query": query})

        if response.status_code == 200:
            data = response.json()
            result = data['result']

            # Create a prompt to inform GPT about the table and its columns
            prompt = "### SQLite Database Table and Columns: "
            prompt += "\n(Table Name: travel_packages Columns: "

            # Format each column name and type
            columns_str = ', '.join([f"{col[0]} ({col[1]})" for col in result])

            prompt += columns_str + ")"

            return prompt
        else:
            return f"Error: {response.status_code} - {response.json()}"
    except Exception as e:
        return str(e)

# Define the API URL and SQL query
api_url = "http://127.0.0.1:5050/query"
query = "SELECT name, type FROM PRAGMA_TABLE_INFO('travel_packages')"

# Call the function with the API URL and query
prompt = create_table_definition_prompt(api_url, query)
print(prompt)
```

```python
import torch
from torch.utils.data import DataLoader, Dataset
from transformers import BertTokenizer, BertForSequenceClassification, AdamW
from torch.nn import CrossEntropyLoss
import pandas as pd
from sklearn.model_selection import train_test_split

# Load your labeled dataset
df = pd.read_csv("labeled_dataset.csv")

# Split the dataset into training and validation sets
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

# Load BERT tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Tokenize the dataset
def tokenize_data(data):
    return tokenizer(data["user_query"].tolist(), truncation=True, padding=True, return_tensors="pt")

train_tokenized = tokenize_data(train_df)
val_tokenized = tokenize_data(val_df)

# Define the intent dataset
class IntentDataset(Dataset):
    def __init__(self, tokenized_data, labels):
        self.tokenized_data = tokenized_data
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {
            "input_ids": self.tokenized_data["input_ids"][idx],
            "attention_mask": self.tokenized_data["attention_mask"][idx],
            "labels": torch.tensor(self.labels[idx]),
        }

# Convert labels to tensor
train_labels = train_df["intent"].map({"general": 0, "database_related": 1}).tolist()
val_labels = val_df["intent"].map({"general": 0, "database_related": 1}).tolist()

# Create DataLoader
train_dataset = IntentDataset(train_tokenized, train_labels)
val_dataset = IntentDataset(val_tokenized, val_labels)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False)

# Initialize BERT model for sequence classification
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)
optimizer = AdamW(model.parameters(), lr=1e-5)

# Move the model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Fine-tune the model
num_epochs = 15
criterion = CrossEntropyLoss()

for epoch in range(num_epochs):
    model.train()
    print(f"epoch ", epoch)
    for batch in train_loader:
        inputs = {key: batch[key].to(device) for key in batch}
        outputs = model(**inputs)
        loss = criterion(outputs.logits, inputs["labels"])
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

# Save the fine-tuned model
model.save_pretrained("fine_tuned_model")
```

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification

# Load the saved fine-tuned model
model = BertForSequenceClassification.from_pretrained("fine_tuned_model")

# Load the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

def classify_intent(text):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors="pt")

    # Make the prediction using the loaded model
    with torch.no_grad():
        outputs = model(**inputs)

    # Get the predicted class
    logits = outputs.logits
    probabilities = torch.nn.functional.softmax(logits, dim=1).detach().numpy()[0]

    # Assuming two classes: 0 for general query, 1 for database-related
    intent_class = 1 if probabilities[1] > probabilities[0] else 0
    return intent_class
```

```python
import openai
from openai import OpenAI

client = OpenAI(
  api_key = 'Your-key'
)

def handle_database_query(nlp_query):
    response = client.completions.create(
        model="text-davinci-003",
        prompt=combine_prompts(nlp_query),
        temperature=0,
        max_tokens=200,
        top_p=1.0,
        frequency_penalty=0.0,
        presence_penalty=0.0,
        stop=["#", ";"]
    )
    query = response.choices[0].text
    if query.startswith(" "):
      query = query.replace('\n', ' ')
      query = "Select"+ query
    return query
```
